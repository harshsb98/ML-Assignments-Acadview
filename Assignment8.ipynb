```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('/home/harsh/Downloads/h.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
cols=['bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated','lat', 'long', 'sqft_living15', 'sqft_lot15']
X=df[cols]
y=df.price
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=70
,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(X_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.coef_
```

```python
y_pred=regressor.predict(X_test)
```

```python
pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
```

```python
from sklearn import metrics
metrics.mean_absolute_error(y_test,y_pred)
```

```python
metrics.mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(metrics.mean_squared_error(y_test,y_pred))
```
